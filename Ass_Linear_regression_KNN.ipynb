```
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import plotly.express as px
link = "https://raw.githubusercontent.com/murpi/wilddata/master/quests/weather2019.csv"
df_weather = pd.read_csv(link)
df_weather['couleur']=pd.DataFrame({'couleur':['r','b','y']})
df_weather['couleur'].fillna(0,inplace=True)
df_weather
```

```
# Your code here :
#from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split

X = df_weather[['MAX_TEMPERATURE_C','MIN_TEMPERATURE_C']]
y = df_weather['SUNHOUR']

X_train,X_test,y_train,y_test=train_test_split(X, y,random_state=1, train_size=0.80)

modelLRKNN = KNeighborsRegressor(n_neighbors=13).fit(X_train ,y_train)
```

```
from sklearn.linear_model import LinearRegression

modelLR = LinearRegression().fit(X_train ,y_train)
```

```
from sklearn.metrics import r2_score

print("Comparaison : \n")

print("Linear regression score with KNN : " ,modelLRKNN.score(X_test, y_test))
print("Linear regression score without KNN : ",modelLR.score(X_test, y_test))
```

```
y_pred=modelLRKNN.predict(X_test)
y_pred1=modelLR.predict(X_test)
print("Linear regression score with KNN : ", r2_score(y_test, y_pred))
print("Linear regression score without KNN : ",r2_score(y_test, y_pred1))
```
