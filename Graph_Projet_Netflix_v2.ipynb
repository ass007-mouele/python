```
# on importe les modules qui nous serviront pour nettoyer nos données, les visualiser et créer notre système de recommandations
import re
import numpy as no
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor, NearestNeighbors 
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler, MultiLabelBinarizer
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error, accuracy_score, classification_report, confusion_matrix

# on importe nos jeux de données
tmdb_credits = pd.read_csv('/content/drive/My Drive/ml-latest/tmdb_5000_credits.csv')
tmdb_movies = pd.read_csv('/content/drive/My Drive/ml-latest/tmdb_5000_movies.csv')

# on fusionne les 2 jeux de données pour pouvoir les retravailler
df_final = tmdb_movies.merge(tmdb_credits, how="left", left_on="id", right_on="movie_id")
```

```
# on supprime les colonnes qui ne nous serviront pas pour l'analyse
df_final.drop(["homepage", "id", "original_language", "tagline", "title_y", "crew"], axis=1, inplace=True)
```

```
# on ne garde que les films qui sont réellement sortis au cinéma
df_final = df_final.loc[df_final["status"] == "Released"]
df_final["release_date"] = pd.to_datetime(df_final["release_date"])
```

```
df_final["release_date"] = df_final["release_date"].astype('str') 
df_final["release_année"]= df_final["release_date"].apply(lambda x: x[0:4])
```

```
# on affiche le jeu de données final
#print(df_final.shape)
df_final.drop([90], inplace=True)
df_finalB=df_final
df_final.groupby('production_countries').mean()
#print(df_final['production_countries'].head(50))
df_final['production_countries']=df_final['production_countries'].apply(lambda x: x[17:19])
df_final['production_countries'].head()
df_final=df_final.groupby('production_countries').mean()
df_final=df_final.reset_index(level=['production_countries'])

```
