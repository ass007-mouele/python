# Mission 1

Modifie le code ci-dessous de manière à ce que :

- La taille de la barre de Bernard s'arrête à 11,5, celle de Martine à 8
- La barre de Jessica doit commencer à 2
- Attribue une couleur différente pour chaque barre

```
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(19680801)


plt.rcdefaults()
fig, ax = plt.subplots()

people = ('Bernard', 'Hector', 'Jessica', 'Jacques', 'Martine')
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))

ax.barh(y_pos, performance, xerr=error, align='center',
        color='green', ecolor='black')
ax.set_yticks(y_pos)
ax.set_yticklabels(people)
ax.invert_yaxis() 
ax.set_xlabel('Avancement')
ax.set_title("Quel est ton taux d'avancement sur les quêtes ?")

plt.show()
```

# Mission 2

Modifie le code ci-dessous dans une nouvelle cellule pour que :
- Les valeurs pour hommes soient à droite et celles pour femmes à gauche
- Modifier la couleur des colonnes
- Diminuer l'épaisseur des paires de baton

```
import matplotlib.pyplot as plt
import numpy as np


labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 34, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]

x = np.arange(len(labels)) 
width = 0.35  

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, men_means, width, label='Men')
rects2 = ax.bar(x + width/2, women_means, width, label='Women')

ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3), 
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
```

# Mission 3

Commente le code ci-dessous pour l'expliquer à un collège

```
import matplotlib
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(19680801)

mu = 100  
sigma = 15  
x = mu + sigma * np.random.randn(437)

num_bins = 50

fig, ax = plt.subplots()

n, bins, patches = ax.hist(x, num_bins, density=1)

y = ((1 / (np.sqrt(2 * np.pi) * sigma)) *
     np.exp(-0.5 * (1 / sigma * (bins - mu))**2))
ax.plot(bins, y, '--')
ax.set_xlabel('Smarts')
ax.set_ylabel('Probability density')
ax.set_title(r'Histogram of IQ: $\mu=100$, $\sigma=15$')

fig.tight_layout()
plt.show()
```
