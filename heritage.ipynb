```
from random import randint
from random import uniform
from random import choice

class Statisticien :
   _logique =randint(0,100)
   _memoire=randint(0,100)
   _creativite=randint(0,100)

   def __init__(self):
      condition=False
      ########################################################################################
      # On rappelle que la somme des trois variables doit etre comprise entre 111 et 177.    #
      # donc toutes les variables doivent etre inferieur ou égal à 177                       #
      # c a d l'intervale de choix des variables est de 0 à 177                              #
      # On crée un tableau des valeurs impaires et non multiples de 5 comprise entre 0 et 177#
      ########################################################################################
      intervale=([i for i in range(0,178) if i%2!=0 and i%5!=0]) 
      somme=self._logique + self._memoire +self._creativite 
      while (somme <111 or somme > 177) or condition==False :
         self._logique=choice(intervale)
         self._memoire=choice(intervale)
         self._creativite=choice(intervale)
         condition=True
         somme=self._logique + self._memoire +self._creativite
      print("logique:{},memoire={},creativite={}".format(self._logique ,self._memoire ,self._creativite))
      print("leur somme={}".format(somme))
      
   def __carrenombre(nombre):
      carre = nombre * nombre # innaccessible par la classe fille #
      return carre

   def __cubenombre(nombre):  # innaccessible par la classe fille #
      cube = nombre **3
      return(cube)

   def _valabs(nombre):
      return abs(nombre)

   def _factoriel(nombre):
      fact=1
      for i in range(1,nombre+1):
         fact=fact*i
      return fact

   def _moyenneliste(listenombre):
      somme=0
      for i in range (len(listenombre)):
         somme+=listenombre[i]
         moyenne=somme/len(listenombre)
      return moyenne

   def _minimum(listenombre):
      mini=listenombre[0]
      for i in range(1,len(listenombre)):
         if listenombre[i] < mini:
            mini=listenombre[i]
      return mini

   def _maximum(listenombre):
      maxi=listenombre[0]
      for i in range(1,len(listenombre)):
         if listenombre[i] > maxi:
            maxi=listenombre[i]
      return maxi

   def _puissanceDeX (number,x):
      return number**x  

   def _nombrepremierinfX(num):
      mesnombrespremier=list()
      for j in range (1,num):
         testeur=0
         for i in range (1,num):
            if j%i==0:
               testeur+=1
         if testeur == 2:     
            mesnombrespremier.append(j)
            testeur=0
      return mesnombrespremier    

   def _fibonnaci():
      liste=[0,1]
      for i in range (2,30):
         num=liste[i-1]+liste[i-2]
         liste.append(num)
      return liste

   def _racinecarre(nombre):
      return nombre**0.5

   def _mmode(liste):
      x=liste.count(liste[0])
      y=liste[0]
      for i in range(1,len(liste)):
         if liste.count(liste[i]) > x:
            x=liste.count(liste[i])
            y=liste[i]
      return y
    
      
   

class developpeur : 
   _frond_end=uniform(0,10)
   _back_end=uniform(0,10)
   _full_stack=randint(0,10) 
   _gestion_de_projet=uniform(0,10)

   def __init__(self):
      self._frond_end=uniform(0,10)
      self._back_end=uniform(0,10)
      self.__gestion_de_projet=uniform(0,10)

   def _vrai_full_stack(self) :
      mmoyenne=int((self._frond_end + self._back_end)//2)
      #################################################################################
      #Tant que full_stack est different de la moyenne on randomise                   #           
      #Dès qu'il est egal a l'entier sup ou inf, on sors de la boucle avec un break   #
      #################################################################################
      while (self._full_stack!=mmoyenne): 
         self._full_stack=randint(0,10) 
         if self._full_stack==(mmoyenne+1) or self._full_stack==(mmoyenne-1) :
            break
      print("la moyenne entre frond_end et back_end est ",mmoyenne)
      print("le vrai full_satck est :",self._full_stack)


class  dataAnalyst(Statisticien,developpeur):
#Classe  data analyste qui herite de deux classes meres :
    nom=str() 
    prenom=str() 
    age=int() 
    sexe=str() 
    formationPrecedente=str() 
    motivation = 100 
    progression=0 

    def __init__(self): # Notre méthode constructeur
        Statisticien.__init__(self)
        developpeur.__init__(self)
        self.nom = 'mouele'
        self.prenom ='assoni'  
        self.age = '38'
        self.formationPrecedente = 'licence'


    def suivreFormation(self):
       self.motivation-=randint(5, 25)
       self.progression+=randint(20, 30)
  
    def bosserPlus(self) : 
       self.motivation+=randint(10, 30)
       self.progression+=randint(10, 30)

    def echouer(self) :
       self.motivation-=randint(20 ,40)
       self.progression+=randint(20 ,40)
 
    def reussir(self) :
       self.motivation+=randint(20 ,40)
       self.progression+=randint(10 ,20)


 

print("Verif héritage de Statisticien:\n", issubclass(dataAnalyst,Statisticien))
print("Verif héritage de developpeur:\n",issubclass(dataAnalyst,developpeur))
print("\n")

assoni=dataAnalyst()
assoni._vrai_full_stack()

```
